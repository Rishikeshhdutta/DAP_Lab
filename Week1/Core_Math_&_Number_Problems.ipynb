```python
n = int(input("Enter an integer: "))

original = abs(n)   # ignore sign for palindrome check
temp = original
rev = 0

while temp > 0:
    rev = rev * 10 + (temp % 10)
    temp //= 10

if original == rev:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
n = int(input("Enter an integer: "))

num = abs(n)      # work with positive value
temp = num

# Step 1: count digits
digits = 0
while temp > 0:
    digits += 1
    temp //= 10

# Step 2: calculate Armstrong sum
temp = num
armstrong_sum = 0

while temp > 0:
    digit = temp % 10
    armstrong_sum += digit ** digits
    temp //= 10

# Step 3: check
if num == armstrong_sum:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
n = int(input("Enter an integer: "))

if n <= 1:
    print("Not a Perfect number")
else:
    divisor_sum = 1   # 1 is always a proper divisor
    i = 2

    while i * i <= n:
        if n % i == 0:
            divisor_sum += i
            if i != n // i:
                divisor_sum += n // i
        i += 1

    if divisor_sum == n:
        print("Perfect number")
    else:
        print("Not a Perfect number")
```

```python
def is_automorphic(num):
    if num < 0:
        return False

    square = num * num

    temp = num
    while temp > 0:
        if temp % 10 != square % 10:
            return False
        temp //= 10
        square //= 10

    return True


num = int(input("Enter a number: "))

if is_automorphic(num):
    print("Automorphic number")
else:
    print("Not an Automorphic number")
```

```python
def is_neon(num):
    if num < 0:
        return False

    square = num * num
    digit_sum = 0

    while square > 0:
        digit_sum += square % 10
        square //= 10

    return digit_sum == num


num = int(input("Enter a number: "))

if is_neon(num):
    print("Neon number")
else:
    print("Not a Neon number")
```

```python
n = int(input("Enter a number: "))

print("Prime factors are:")

# Check for factor 2
while n % 2 == 0:
    print(2, end=" ")
    n //= 2

# Check for odd factors
i = 3
while i * i <= n:
    while n % i == 0:
        print(i, end=" ")
        n //= i
    i += 2

# If remaining n is prime
if n > 1:
    print(n)
```

```python
n = int(input("Enter the upper limit: "))

# Step 1: Assume all numbers are prime
is_prime = [True] * (n + 1)
is_prime[0] = is_prime[1] = False

# Step 2: Mark multiples as non-prime
i = 2
while i * i <= n:
    if is_prime[i]:
        for j in range(i * i, n + 1, i):
            is_prime[j] = False
    i += 1

# Step 3: Print primes
print("Prime numbers are:")
for i in range(2, n + 1):
    if is_prime[i]:
        print(i, end=" ")
```

```python
n = int(input("Enter the upper limit: "))

# Step 1: Sieve of Eratosthenes
is_prime = [True] * (n + 1)
is_prime[0] = is_prime[1] = False

for i in range(2, int(n**0.5) + 1):
    if is_prime[i]:
        for j in range(i * i, n + 1, i):
            is_prime[j] = False

# Step 2: Find twin primes
print("Twin primes are:")
for i in range(3, n - 1):
    if is_prime[i] and is_prime[i + 2]:
        print(f"({i}, {i + 2})", end=" ")
```

```python
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    return n * factorial(n - 1)

n = int(input("Enter a number: "))
print("Factorial:", factorial(n))
```

```python
def factorial(n):
    res = 1
    for i in range(2, n + 1):
        res *= i
    return res

n = int(input("Enter n: "))
r = int(input("Enter r: "))

# nCr is not defined if r > n
if r > n:
    print("Invalid input")
else:
    r = min(r, n - r)   # simplification
    ncr = factorial(n) // (factorial(r) * factorial(n - r))
    print("nCr =", ncr)
```

```python
import math

x = float(input("Enter x (in degrees): "))
n = int(input("Enter number of terms: "))

# Convert degrees to radians
x = math.radians(x)

sinx = 0
sign = 1
power = x
fact = 1

for i in range(1, 2*n, 2):
    sinx += sign * power / fact
    sign *= -1
    power *= x * x
    fact *= (i + 1) * (i + 2)

print("sin(x) =", sinx)
```
